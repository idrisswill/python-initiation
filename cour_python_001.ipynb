## Mise en evidence d'un tableau contigu en mémoire (un bloc continu de cases contenant des références).


```python
import time

n = 100_000_000

lst = list(range(n))

# insertion au début
start = time.time()
lst.insert(0, -1)
end = time.time()

print("Insertion début :", end - start)

# insertion à la fin
lst = list(range(n))

start = time.time()
lst.append(-1)
end = time.time()

print("Insertion fin :", end - start)
```

- Si dit un nombre qui s'ecrit avec un nombre de chiffres pairs, il faut remplacée par deux nombre. La moitié gauche des chiffres est appart, et la moitié droite est autre nombre. (Les nouveaux chiffres ne gardent pas de zéros en tête : 1000 deviendrait des  10 et 0.)

example:

Si il dispose des chiffres 0 1 10 99 999 et qu'il cligne des yeux une fois, tu auras :

Le deuxième nombre, 1, est multipliée par 2024 pour devenir 2024.
Le troisième nombre, 10, est divisée en un nombre 1 suivie d’un autre 0.
Le quatrième nombre, 99, est divisée en deux nombres marquées 9.
Le cinquième nombre, 999, est remplacé par 999 * 2024 = 2021976.

resultat: 1 2024 1 0 9 9 2021976.

😵‍💫 Le cas de M. Tchakouté

Maintenant que M. Tchakouté est quelque peu ivre, il a quitté par inadvertance la plateforme digitale de sa banque. Pour y retourner, il lui faut un mot de passe.

Prévenant comme toujours, il avait noté ce mot de passe dans un fichier `.txt` sur son ordinateur.

Petit problème : le mot de passe saisi par M. Tchakouté ne permet pas de déverrouiller sa banque. Tu te proposes alors de l’aider.

Bien qu’il soit un peu éméché, il n’est pas idiot pour autant : il refuse de te donner le fichier et préfère te le dicter les valeurs.

Après trois essais, la plateforme reste toujours verrouillée. En l’observant attentivement, tu remarques qu’à chaque fois qu’il cligne des yeux, il lit une suite de chiffres différente.

Malin comme tu es, et soucieux de lui venir en aide, tu identifies un pattern :

---

🔁 Règles de transformation :

- Si le nombre est **0**, il devient **1**.
- Si le nombre possède un **nombre pair de chiffres**, il est remplacé par deux nombres :
  - la moitié gauche des chiffres
  - la moitié droite des chiffres
  *(sans zéros non significatifs : 1000 devient 10 et 0)*
- Dans tous les autres cas, le nombre est simplement multiplié par **2024**.

---

📌 Exemple :

Si la suite initiale est :
0 1 10 99 999

Après un clignement d’yeux, on obtient :

- 0 → 1
- 1 → 2024
- 10 → 1 0
- 99 → 9 9
- 999 → 999 × 2024 = 2021976

---

✅ Résultat :
1 2024 1 0 9 9 2021976

Le mot de passe correspond au nombre de blocs finaux, soit ici : **7**.

---

# Fonction recursive

Une fonction récursive est une fonction qui s’appelle elle-même.

“Je ne sais pas tout résoudre d’un coup… donc je vais me rappeler de moi-même pour avancer 😄”

```python
def somme(n):
    if n == 0:
        return 0
    return n + somme(n - 1)

print(somme(5))
```

```python
import time
def compte_a_rebours(n):
    # cas de base
    if n == 0:
        print("Fin 🚀")
        return

    # affichage
    print(n)
    time.sleep(1)

    compte_a_rebours(n - 1)

compte_a_rebours(5)
```
